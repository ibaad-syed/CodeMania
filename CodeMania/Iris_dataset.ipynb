```python
null=0
def distance (x1,y1,z1,l1,x2,y2,z2,l2):
    xn = x1-x2
    if(xn.any()<null):
        xn=xn*(-1)
    yn= y1-y2
    if (yn.any()<null):
        yn=yn*(-1)
    zn= z1-z2
    if (zn.any()<null):
        zn=zn*(-1)
    ln= l1-l2
    if (ln.any()<null):
        ln=ln*(-1)
    return xn+yn+zn+ln
```
