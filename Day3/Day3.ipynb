## Comprehension Methods

### List Comprehension
`a=[expr for item in list]`
### Extended List Comprehension
`a = [expr for item in list if condition]`
### Dictionary Comprehension
`a = {expr:expr for item in list}`<br>
`a = {expr:expr for item in list if condition}`

```python
[x for x in range(10)]
```

```python
[x**2 for x in range(10)]
```

```python
[x**2 for x in range(10) if x%2 != 0]
```

```python
def f(x):
    return x*x*x - x*x

[f(x) for x in range(10)]
```

## [Josephus Problem](https://en.wikipedia.org/wiki/Josephus_problem)

```python
# Nannaku Prematho Problem
N = 100
alive = list(range(N)) # initially all N people are alive
flag = 0
while len(alive) > 1: # if only one is alive exit loop
    print(alive)
    last = alive[-1]
    alive = [x for x in alive if alive.index(x) % 2 == flag] # 'kill' alternate people from the list 
    # ,starting from the first or second index based on whether the last person in the list is alive or killed.
    # if last person is killed ,then start 'killing' people from the 2nd position(index 1) of list.
    # if last person is not killed , then start 'killing' from the 1st position(index = 0) of list
    
    if last in alive:
        flag = 1 
    else:
        flag = 0
    
print(alive)
```
